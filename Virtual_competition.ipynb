#**Diabetes Dataset**

````python
import seaborn as sns
````

````python
dm=pd.read_csv('diabetes_model.csv')
dm.head()
````

#Univariate Analysis:
Analyze individual variables (e.g., histograms, box plots).


````python
dm.hist(figsize=(10,10))
plt.show()
````

````python
dm.plot(figsize=(13,6))
plt.show()
````

#Bivariate Analysis:

Explore relationships between pairs of variables (e.g., scatter plots, correlation matrices).

````python
x=(dm['Pregnancies'])
y=(dm['Age'])
plt.scatter(x,y,color='red')
plt.xlabel('Pregnancies')
plt.ylabel('Age')
plt.title('Pregnancies vs Age')
plt.show()
````

````python
dm[['Pregnancies','Age']].corr()
````

#Multivariate Analysis:
Investigate interactions among multiple variables (e.g., pair plots, 3D plots).

````python
plt.figure(figsize=(10,10))
sns.pairplot(dm[['Age','BMI','Pregnancies','Glucose']])
plt.show()
````

#Insights

*   Older age is correlated with a high probability of diabaties.
*   Higher number of pregnancies is weakly associated with the
  likelihood of diabeties.
*   Higher BMI is linked with an increased chance of diabeties.
*   Insulin,blood pressure, skin thickness shows weaker correleations
  with the outcome.
